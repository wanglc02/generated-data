```python
import os
import json
dataset_dir = '/home/shixianjie/workspace/Projects/generated-data/dataset/datasets/SQuAD2.0/all_passages/QA_tune'

def convert_to_json(split, output_file):
    questions_path = os.path.join(dataset_dir, f'{split}.source')
    answers_path = os.path.join(dataset_dir, f'{split}.target')
    
    # 读取问题和答案
    with open(questions_path, 'r', encoding='utf-8') as q_file:
        questions = q_file.readlines()
    
    with open(answers_path, 'r', encoding='utf-8') as a_file:
        answers = a_file.readlines()
    
    # 确保问题和答案数量一致
    assert len(questions) == len(answers), f"Mismatch in number of questions and answers in {dataset_dir}"
    
    # 构建数据列表
    data = []
    for q, a in zip(questions, answers):
        q = q.strip()
        a = a.strip()
        if q and a:  # 过滤空行
            data.append({
                "source": q,
                "target": a
            })
    
    # 保存为 JSON 文件
    with open(output_file, 'w', encoding='utf-8') as out_file:
        for entry in data:
            json_line = json.dumps(entry, ensure_ascii=False)
            out_file.write(json_line + '\n')
    
    print(f"Saved {len(data)} entries to {output_file}")

def main():
   
    output_dir = '../dataset/squadqa'
    
    os.makedirs(output_dir, exist_ok=True)
    
    splits = ['train', 'test', 'dev']
    
    for split in splits:
        split_dir = os.path.join(dataset_dir, split)
        output_file = os.path.join(output_dir, f"{split}.json")
        convert_to_json(split, output_file)

if __name__ == "__main__":
    main()
```
